```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("Sheet1.csv")
data.head()
```

```python
time_stamp = "3/26/2020 21:35:33"
person_name = ["Bobby Brown", "Dolly Hunter", "Edna White", "Gertrude Smith", "Irving Stafford", "John Hifford"]
person_email = ["personal_"+str(i) for i in range(4)]
name = ["BusinessName"+str(i) for i in range(4)]
phone = "8001201500"
bus_email = ["business_"+str(i) for i in range(4)]
type_bus = [["Bakery", "Cafe"], ["Book Store", "Art/Jewelry"], ["Restaurant", "Bakery"]]
state_city = {"Illinois - IL": ["Chicago", '60007'], "California - CA": ["Los Angeles", '90037'], "New York - NY": ["New York City", '10001']}
street = "1200 Main Street"
site = "www.mysite.com"
commerce = "www.mycommerce.com"
fb = "www.facebook.com/mypage"
twitter = "www.twitter.com/mytwitter"
instagram = "www.instagram.com/myinsta"
delivery_text = "Online orders here: [link]"
help_text = "We would appreciate any purchases of small items."
gift_card = ["Yes, in store.", "Yes, online.", "No."]
```

```python
entries = []
for s in state_city.keys():
    for a in person_name:
        for b in person_email:
            for n in name:
                for e in bus_email:
                    for t in type_bus:
                        for maint in t:
                            for g in gift_card:
                                entries.append({"Timestamp": time_stamp, 'PersonName': a, 'PersonEmail': b+"@site.com", 'Name': n, 'Phone': phone, 'Email': e+"@site.com",
           'Description': '', 'Type': t[0]+", "+t[1], 'MainType': maint, 'State': s, 'City': state_city[s][0], 'Street': street, 'ZIP': state_city[s][1],
           'Site': site, 'Commerce': commerce, 'FB': fb, 'Twitter': twitter, 'Instagram': instagram, 'DeliveryText': delivery_text,
           'HelpText': help_text, 'GiftCard': g, 'GiftCardLink': ''})                      
```

```python
for entry in entries:
    data = data.append(entry, ignore_index=True)
```

```python
data.to_csv("out.csv", index=False)
```
